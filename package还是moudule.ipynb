Library（库）
└── Package（包）
    ├── Subpackage（子包）
    └── Module（模块）
        ├── Function（函数）
        ├── Class（类）
        ├── Variable（变量）
        └── Object（对象）

package: 一个包含\_\_init__.py的文件夹\
module: 一个.py文件，包含function, class, variable等

# 检查是Packages还是modules
package有__path__属性

```python
import numpy as np
import matplotlib.pyplot as plt
print(hasattr(np, '__path__')) # True → 是 package
print(hasattr(plt, '__path__')) # False → 是 module（普通 .py 文件）

#查看路径
print(np.__file__)
print(plt.__file__)
```

package: \_\_file__通常指向\_\_init__.py
module: \_\_file__通常指向具体的.py文件
